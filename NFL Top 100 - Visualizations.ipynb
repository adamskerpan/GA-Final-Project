```python
df = pd.read_csv('nfl_top_100.csv')
df[df['Pos'] == 'DT']
```

```python
AaronDonald_rank = df[df['Player'] == 'Aaron Donald']['Rank']
AaronDonald_year = df[df['Player'] == 'Aaron Donald']['Year']

Suh_rank = df[df['Player'] == 'Ndamukong Suh']['Rank']
Suh_year = df[df['Player'] == 'Ndamukong Suh']['Year']

Atkins_rank = df[df['Player'] == 'Geno Atkins']['Rank']
Atkins_year = df[df['Player'] == 'Geno Atkins']['Year']
```

```python
plt.plot(AaronDonald_year, AaronDonald_rank, label = "Aaron Donald", color='#4ca832', marker='o')
plt.plot(Suh_year, Suh_rank, label = "Ndamukong Suh", color='b', marker='.', linestyle='--')
plt.plot(Atkins_year, Atkins_rank, label = "Geno Atkins", color='red', marker='.', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Rank")
plt.title("Aaron Donald vs. Ndamukong Suh vs. Geno Atkins")
plt.legend()
plt.grid(True)
plt.show()
```
