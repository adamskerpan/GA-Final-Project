```python
# Modules to scrape the URL
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 2019 Data

url2019 = 'https://www.pro-football-reference.com/awards/2019-nfl-top-100.htm'
html2019 = urlopen(url2019)
stats_page_2019 = bs(html2019)

column_headers_2019 = stats_page_2019.findAll('tr')[1]
column_headers_2019 = [i.getText() for i in column_headers_2019.findAll('th')]
print(column_headers_2019)
```

```python
rows_2019 = stats_page_2019.findAll('tr')[2:]
```

```python
stats_2019 = []
for i in range(len(rows_2019)):
    stats_2019.append([col.getText() for col in rows_2019[i].findAll('td')])
```

```python
df_2019 = pd.DataFrame(stats_2019, columns=column_headers_2019[1:])

# Add what year it is for when we combine the data frames
df_2019.insert(0, column="Year", value="2019")

# Add the player's rank to the data frame
df_2019.insert(1, column="Rank", value = df_2019.index +1)
```
